```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

def show_data_figure(input_data, output_data):
    plt.figure(figsize=(15, 6))
    plt.plot(input_data, color='blue', label='Input Data', marker='o', markersize=2)
    plt.plot(output_data, color='green', label='Output Data', marker='o', markersize=2)

    for i in range(len(input_data)):
        line_color = 'red' if abs((input_data[i] - output_data[i])[0]) > 5 else 'green'
        plt.plot([i, i], [input_data[i], output_data[i]], color=line_color, linewidth=1)

    plt.legend()
    plt.show()


# Create input data
sample_amount = 10
num_range = 1000
length_between = 5

input_data = np.random.randint(0, num_range, size=(sample_amount, 1))

# Randomize the addition, 50% positive, 50% negative
random_1_to_5_to_add = np.random.choice([-1, 1], size=(sample_amount, 1)) * np.random.randint(1, length_between, size=(sample_amount, 1))
output_data = input_data + random_1_to_5_to_add

# Sort input data and corresponding output data
sorted_indices = np.argsort(input_data.flatten())
input_data = input_data[sorted_indices]
output_data = output_data[sorted_indices]

# Visualize input output
show_data_figure(input_data, output_data)
```
